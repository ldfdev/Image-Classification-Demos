Data has to be splitted into train, test subsets

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(np.random.randint(20, 56, (40, 5)),columns=['A','B','C','D','E'])
data
```

```python
def randomSplitter(dataset_size, random_seed=42, margin=0.2):
    '''
        randomly shuffles the range(0, dataset_size)
        and return 2 slices, separated according to margin from the dataset_size
    '''
    indices = list(range(dataset_size))
    split_index = int(np.floor(margin * dataset_size))
    np.random.seed(random_seed)
    np.random.shuffle(indices)
    train_indices, val_indices = indices[split_index:], indices[:split_index]
    return train_indices, val_indices
```

```python
train_set_indices, test_set_indices = randomSplitter(data.shape[0], margin=0.4)
```

```python
new_data = data.iloc[train_set_indices].reset_index(drop=True)
new_data.columns
# .drop(columns=['Index'])
new_data.iloc[5][2:]
new_data
# indexing as this is useful implementing __getitem__()
```
