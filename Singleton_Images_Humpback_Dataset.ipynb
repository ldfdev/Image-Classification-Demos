```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import torch, torchvision
import PIL
```

```python
original_data = pd.read_csv(os.path.join('../data/', 'train.csv'))
original_data.columns = ['Image', 'Whale_ID']
original_data = original_data[original_data.Whale_ID != 'new_whale']
```

```python
original_data.head(4)
```

```python
singletons = pd.concat([original_data.Whale_ID.value_counts()],axis=1).reset_index()
type(singletons)
singletons.columns=['Whale_ID', 'Frequency_Whale']
singletons = singletons[singletons.Frequency_Whale == 1]
singletons
```

```python
singletons.shape
```

#### Original dataFrame and the one containing Singletons should be merged to get corresponding image filename

```python
singletons_named_files = pd.merge(original_data,
                                 singletons,
                                 left_on='Whale_ID',
                                 right_on='Whale_ID',
                                 how='inner') # inner join
singletons_named_files.head(4)
```

```python
singletons_named_files.shape
```

```python
singletons_named_files = singletons_named_files.drop(columns=['Frequency_Whale'])
singletons_named_files.head(4)
```
