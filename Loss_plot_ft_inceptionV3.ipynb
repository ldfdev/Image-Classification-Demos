```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (20, 12)
%matplotlib inline
```

```python
data=pd.read_csv(os.path.join('../dataset_statistics', 'stats_fine_tuning_inception_v3.csv'))
data.head()
print(data.columns)
```

```python
plt.plot(data.Epoch, data.loc[:,(' Train Loss')], '-r', label='Train Loss')
plt.plot(data.Epoch, data.loc[:,(' Acc_train')], '--g', label='Acc_train')
plt.plot(data.Epoch, data.loc[:,(' val Loss')], '-.b', label='val Loss')
plt.plot(data.Epoch, data.loc[:,(' Acc_val')], 'oy', label='Acc_val')
plt.legend(loc='best')
# plt.show()

plt.savefig('Epoch_vs_loss_ft_inceptionv3.png', bbox_inches='tight')
```
